```python
import pandas as pd
```

```python
all_df=pd.DataFrame(columns=["time","vm","cpu_demand"])
```

```python
df = pd.read_csv("2023-07-01_2023-07-31_vm_cpu_demandPct_Spain_OCS_part002.csv")
```

```python
df
```

```python
columns = df.columns.to_list()
columns
```

```python

for items in columns[1:]:
    df_new = df[['time',items]]
    df_new =df_new.assign(vm=items)
    df_new.rename(columns={items:"cpu_demand"},inplace=True)
    all_df=all_df.append(df_new)
```

```python
all_df.shape
```

```python
all_df.to_csv("cpu_demand_es_ocs_june_july.csv",index=False)
```
